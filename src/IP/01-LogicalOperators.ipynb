***Logic Operators***

```
#import libraries
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```
# Import images from google drive
from google.colab import drive
drive.mount('/content/drive/')
```

```
#read the image
# reading image and converting to gray scale
img = Image.open('/content/drive/MyDrive/College/5th-Sem/IP/Images/aa.png').convert('L')
# display image
img
```

```
img = img.resize((400,400), Image.ANTIALIAS)
# convert to numpy array
numpy_image = np.array(img)
```

```
#read the image
# reading image and converting to gray scale
img2 = Image.open('/content/drive/MyDrive/College/5th-Sem/IP/Images/bb.png').convert('L')
# display image
img2
```

```
img2 = img2.resize((400,400), Image.ANTIALIAS)
# convert to numpy array
numpy_image2 = np.array(img2)
```

```
# Logic AND
aa = numpy_image*numpy_image2
aa = Image.fromarray(aa)
aa = aa.convert("L")
aa
```

```
# Logic OR
aa = numpy_image+numpy_image2
aa = Image.fromarray(aa)
aa = aa.convert("L")
aa
```

```
#XOR
aa = np.logical_xor(numpy_image, numpy_image2)
aa = Image.fromarray(aa)
aa = aa.convert("L")
aa
```
